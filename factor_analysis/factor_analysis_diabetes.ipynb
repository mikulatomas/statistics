```python
from sklearn.decomposition import FactorAnalysis
from datasets import load_diabetes
import plotly.express as px
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
df_original, df_train, df_test = load_diabetes()
```

```python
train_X = df_train.drop(['target'], axis=1)
train_y = df_train['target']
test_X = df_test.drop(['target'], axis=1)
test_y = df_test['target']
original_X = df_original.drop(['target'], axis=1)
original_y = df_original['target']
features = train_X.columns
```

```python
train_X.shape
```

```python
transformer = FactorAnalysis(n_components=3, random_state=0, rotation=None).fit(train_X)
```

```python
transformed_X = transformer.transform(train_X)
```

```python
transformed_df = pd.DataFrame(transformed_X, columns=['x','y','z'])
transformed_df['target'] = train_y.values
# transformed_df['name'] = df.index
transformed_df['size'] = [1 for i in range(transformed_df.shape[0])]
px.scatter_3d(transformed_df, x='x', y='y', z='z',
              color='target', size='size', size_max=15)
```

```python
transformer = FactorAnalysis(n_components=3, random_state=0, rotation='varimax').fit(train_X)
```

```python
transformed_X = transformer.transform(train_X)
```
