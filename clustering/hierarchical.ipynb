```python
from datasets import load_dutch
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import MinMaxScaler
```

```python
import numpy as np

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering


def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
df = load_dutch()
```

```python
train_X = df.drop('target', axis=1)
train_y = df['target']

scaler = MinMaxScaler().fit(train_X)
train_X = scaler.transform(train_X)
```

```python
train_X.shape
```

```python
# kde dendogram riznout abychom odhadli k?
```

```python
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None, linkage='complete').fit(train_X)
```

```python
plot_dendrogram(model)
```

```python
# No more than p levels of the dendrogram tree are displayed. A level includes all nodes with p merges from the last merge.
plot_dendrogram(model, truncate_mode='level', p=4)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
```
